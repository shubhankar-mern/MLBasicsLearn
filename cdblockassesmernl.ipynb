```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
Y = (X-5)**2
print('X ==',X)
print('Y ==',Y)
plt.plot(X,Y)
plt.show()
```

```python
x = 2
lr = 0.1
error = []
plt.plot(X,Y)

for i in range(150):
  grad = 2*(x - 5)
  x = x - lr*grad
  y = (x-5)**2
  error.append(y)
  plt.scatter(x,y)
  print(x)
print(error)
```

```python
## surface plot
## a will be repelted along row b number of times
## b will be repeated along col a number of times
## now u can geberate a coordinate in cartesina plane
```

```python
a = np.arange(-1,1,0.02)
b = a

a,b = np.meshgrid(a,b)
print(a)
print(b)
```

```python
from matplotlib import projections
fig = plt.figure()
axes = fig.add_subplot(projection = '3d')
axes.plot_surface(a,b,a**2+b**2,cmap='rainbow')
plt.show()
## cmap can be rainbow or coolmap
```

```python
## contour plot|

fig = plt.figure()
axes = fig.add_subplot(projection = '3d')
axes.contour(a,b,a**2+b**2,cmap='rainbow')
plt.show()
## cmap can be rainbow or coolmap
```
