```python
df = pd.read_csv('/content/content/MyDrive/titanic.csv')
df.head()
```

```python
df_new = df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
df_new.head()
```

```python
df_new.isna().sum()
```

```python
import math
median_age = math.floor(df_new.Age.median())
median_age
```

```python
df_new.Age = df_new.Age.fillna(median_age)
```

```python
df_new.isna().sum()
```

```python
df_new.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
sex_n = LabelEncoder()
```

```python
df_new['sexx'] = sex_n.fit_transform(df['Sex'])
```

```python
df_new.head()
```

```python
df1 = df_new.drop(['Sex'],axis='columns')
df1.head()
```

```python
from sklearn import tree
```

```python
inputs = df1.drop(['Survived'],axis='columns')
outputs = df1['Survived']
```

```python
inputs
```

```python
outputs
```

```python
from sklearn.model_selection import train_test_split
```

```python

X_train,X_test,y_train,y_test = train_test_split(inputs,outputs, train_size=0.2)
```

```python
y_test
```

```python
dtmod = tree.DecisionTreeClassifier(criterion='entropy')
```

```python
dtmod.fit(X_train,y_train)
```

```python
dtmod.score(X_test,y_test)
```
