```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(10)
y1 = x**2
y2 = 2*x+3
print(x)
print(y1)
print(y2)
```

```python
plt.plot(x,y1,color='red',label='a',marker='o')
plt.plot(x,y2,color='green',label='b',linestyle='dashed')
plt.xlabel("x")
plt.ylabel('y')
plt.title('sum')
plt.legend()
plt.show()
```

```python
##as u dont provide 2 indices becomes x and provided y
plt.plot(y1)
plt.show()
```

```python
##scatter plots
```

```python
plt.figure(figsize=(5,5))
plt.scatter(x,y1,color='red',label='a',marker='^')
plt.scatter(x,y2,color='green',label='b',linestyle='dashed')
plt.xlabel("x")
plt.ylabel('y')
plt.title('sum')
plt.legend()
plt.show()
```

```python
x=np.array([0,1,2])*2
plt.bar(x,[10,20,50], width= 0.5, label="cutrrent",color='yellow' ,tick_label=['a','b','c'])
plt.bar(x+0.5,[20,10,15], width=0.5, label="nxt",color='orange')
plt.xlabel('metal')
plt.ylabel('price')
plt.ylim(0,70)
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/movie_metadata.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
titles = list(df.get('movie_title'))
titles[:5]
```

```python
freq_title = {}
for title in titles:
  length = len(title)

  if freq_title.get(length) is None:
    freq_title[length] = 1
  else:
    freq_title[length] += 1  
```

```python
freq_title
```

```python
X = np.array(list(freq_title.keys()))
```

```python
Y = np.array(list(freq_title.values()))
```

```python
X.shape
```

```python
plt.scatter(X,Y)
plt.xlabel("title")
plt.ylabel("freqcountof movies having length")
plt.title("movie visualtioz")
plt.show()
```
